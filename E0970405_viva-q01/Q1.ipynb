#### Create a function get_genome(person_id)

```python
def get_genome(person_id):
  with open(f'person_{person_id:05d}.txt', 'r') as file:
    file_content = file.read()
  return file_content
```

#### Estimate the infection rate in the population

```python
import numpy as np

num_people = 100
num_infected = 0

for i in range(num_people):
    person_id = np.random.randint(0, 499)
    genome = get_genome(person_id)
    if test(genome) == True:
        num_infected += 1

infection_rate = (num_infected / num_people) * 100
print(f"Infection rate: {infection_rate}%")
```

#### Write a function pool(list_of_id)

```python
def pool(list_of_id):
    genomes = []
    for i in list_of_id:
        genome = get_genome(i)
        genomes.append(genome)
    combined_genome = "".join(genomes)
    return combined_genome
```

#### Apply test() to combined genomes of 20 and 21 and 21 and 22

```python
combined_20_21 = pool([20, 21])

test(combined_20_21)
```

```python
combined_21_22 = pool([21, 22])

test(combined_21_22)
```

Results are expected, since persons 20, 21 and 22 were originally not infected. Hence, joining their genomes / mixing their blood samples would not create infections.

#### For the first 100 persons, test in groups of 10 to determine those infected.

```python
persons = list(range(100))
groups = [persons[i:i+10] for i in range(0, 100, 10)]
infected_list = []
test_counter = 0

for i, group in enumerate(groups):
    combined_genome = pool(group)
    group_result = test(combined_genome)
    test_counter += 1
    
    if group_result == True:
        for person in range(10):
            person_genome = get_genome(groups[i][person])
            test_result = test(person_genome)
            test_counter += 1
            
            if test_result == True:
                infected_list.append(groups[i][person])

print(f"List of infected persons: {infected_list}")
print(f"Number of tests performed: {test_counter}")
```

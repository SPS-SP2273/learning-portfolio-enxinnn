## 1 A Bucket of water

### 1.1 Getting a feel

```python
from matplotlib import pyplot as plt
```

```python
max_time = 5                     # Maximum time (mins)
dt = .5                          # Time step (mins)
rate = 5                         # Rate (L/min)
all_volume = []                  # To keep track of all volumes
volume = 0                       # Starting volume

all_time = np.arange(start=0, stop=max_time, step=dt)  # np.arange returns evenly spaced values within given interval

for time in all_time:
    all_volume.append(volume)    # Record volume
    dV = rate * dt               # Calculate change in volume
    volume += dV                 # Update the new volume

plt.plot(all_time, all_volume)
plt.ylabel('Volume(L)')
plt.xlabel('Time(mins)')
```

### 1.2 An overflowing bucket

```python
max_time = 5                     # Maximum time (mins)
dt = .5                          # Time step (mins)
rate = 5                         # Rate (L/min)
bucket_capacity = 9              # L
all_volume = []                  # To keep track of all volumes
volume = 0                       # Starting volume

all_time = np.arange(start=0, stop=max_time, step=dt)

for time in all_time:
    all_volume.append(volume)
    dV = rate * dt
    if volume <= bucket_capacity:
        volume += dV             # Update the new volume

plt.plot(all_time, all_volume)
plt.ylabel('Volume(L)')
plt.xlabel('Time(mins)')
plt.hlines(bucket_capacity, 0, max_time, colors='grey', ls='dashed')
```

```python
max_time = 5                     # Maximum time (mins)
dt = .5                          # Time step (mins)
rate = 5                         # Rate (L/min)
bucket_capacity = 9              # L

for dt in [0.5, 0.1, 0.05, 0.01, 0.001]:
    all_volume = []                  # To keep track of all volumes
    volume = 0                       # Starting volume

    all_time = np.arange(start=0, stop=max_time, step=dt)

    for time in all_time:
        all_volume.append(volume)
        dV = rate * dt
        if volume <= bucket_capacity:
            volume += dV

    plt.plot(all_time, all_volume, label=f'dt={dt}')

plt.ylabel('Volume(L)')
plt.xlabel('Time(mins)')
plt.hlines(bucket_capacity, 0, max_time, colors='grey', ls='dashed')
plt.legend()
```

### 1.3 A leaky bucket

```python
max_time = 5                     # Maximum time (mins)
dt = .001                        # Time step (mins)
rate = 5                         # Filling rate (L/min)
leak_rate = 1.5                  # L/min
bucket_capacity = 9              # L
all_volume = []                  # To keep track of all volumes
volume = 0                       # Starting volume

all_time = np.arange(start=0, stop=max_time, step=dt)

for time in all_time:
    all_volume.append(volume)

    dV = rate * dt
    leak_volume = leak_rate * dt
    volume -= leak_volume

    if volume <= bucket_capacity:
        volume += dV             # Update the new volume

plt.plot(all_time, all_volume)
plt.ylabel('Volume(L)')
plt.xlabel('Time(mins)')
plt.hlines(bucket_capacity, 0, max_time, colors='grey', ls='dashed')
```

### 1.4 Let’s turn off the tap

```python
tap_off_time = 3                 # When the tap goes off
max_time = 5                     # Maximum time (mins)
dt = .001                        # Time step (mins)
rate = 5                         # Filling rate (L/min)
leak_rate = 1.5                  # L/min
bucket_capacity = 9              # L
all_volume = []                  # To keep track of all volumes
volume = 0                       # Starting volume

all_time = np.arange(start=0, stop=max_time, step=dt)

for time in all_time:
    all_volume.append(volume)
    dV = rate * dt

    leak_volume = leak_rate * dt
    volume -= leak_volume

    if (volume <= bucket_capacity) and (time < tap_off_time):
        volume += dV             

plt.plot(all_time, all_volume)
plt.ylabel('Volume(L)')
plt.xlabel('Time(mins)')
plt.hlines(bucket_capacity, 0, max_time, colors='grey', ls='dashed')
```

### Using while

```python
tap_off_time = 3
max_time = 5
dt = .001
rate = 5
leak_rate = 1.5
bucket_capacity = 9
all_volume = []
all_time = []
volume = 0
time = 0

while time <= max_time:
    all_time.append(time)
    all_volume.append(volume)
    dV = rate * dt
    
    leak_volume = leak_rate * dt
    volume -= leak_volume
    
    if (volume <= bucket_capacity) and (time < tap_off_time):
        volume += dV
    
    time += dt

plt.plot(all_time, all_volume)
plt.ylabel('Volume(L)')
plt.xlabel('Time(mins)')
plt.hlines(bucket_capacity, 0, max_time, colors = 'grey', ls = 'dashed')
```

```python
decay_constant = 142
stop_fraction = 1E-3

N0 = 1
dt = .001
time, N = 0, N0

all_N, all_time = [], []

while True:                 # keeps running until break
    all_time.append(time)
    all_N.append(N)
    
    dN = -decay_constant*N*dt
    N += dN
    
    if N < N0*stop_fraction:
        break
    
    time += dt
    
plt.plot(all_time, all_N)
plt.ylabel('$N$')
plt.xlabel('Time (Millions of years)')
```

## Exercise 1 :  A more realistic leak

```python
def leak_rate(volume):
    volume = volume/1000
    height = volume / (700*10E-4)
    speed = np.sqrt(2*9.8*height)
    area = 1E-4
    leak_rate = area*speed
    return leak_rate

tap_off_time = 3
max_time = 5
dt = .001
rate = 5
bucket_capacity = 9
all_volume, all_time = [], []
volume = 0
time = 0

while time <= max_time:
    all_time.append(time)
    all_volume.append(volume)
    dV = rate * dt
    
    leak_volume = leak_rate(volume)*1000 * dt
    volume -= leak_volume
    
    if (volume <= bucket_capacity) and (time < tap_off_time):
        volume += dV
        
    time += dt
    
plt.plot(all_time, all_volume)
plt.xlabel('Time(mins)')
plt.ylabel('Volume(L)')
plt.show()
```
